# a) Demostrar que la regla de actualización de pesos en la Ecuación (1.3) se mueve en la dirección de clasificar correctamente x(t).

Se puede examinar la regla de actualización de pesos y ver cómo afecta la clasificación de x(t).

La regla de actualización de pesos se expresa como:
w(t+1) = w(t) + y(t) * x(t)

Suponiendo que x(t) está clasificado incorrectamente por w(t), lo que significa que y(t) * w(t) * x(t) ≤ 0. Esto implicaria que la predicción actual de y(t) sea incorrecta.

Ahora considerando el vector de pesos actualizado w(t+1) y su producto escalar con x(t):

w(t+1) * x(t) = (w(t) + y(t) * x(t)) * x(t)
= w(t) * x(t) + y(t) * x(t) * x(t)

Dado que x(t) * x(t) siempre es positivo, tenemos:
w(t+1) * x(t) = w(t) * x(t) + y(t) * ||x(t)||^2

Enfocandonos en y(t) * ||x(t)||^2, se evidencia que representa un factor de escala positivo de x(t) basado en su norma al cuadrado, este término afecta la dirección de la actualización de pesos.

Si x(t) está clasificado incorrectamente por w(t), significa que y(t) * w(t) * x(t) ≤ 0, por ende el término de actualización y(t) * ||x(t)||^2 será positivo y asi el vector de pesos se moverá en la dirección de x(t), intentando corregir la clasificación incorrecta.

Se concluye que la regla de actualización de pesos en la Ecuación (1.3)  se mueve en la dirección de clasificar correctamente x(t) cuando x(t) está clasificado incorrectamente por w(t), como se muestra arriba.
