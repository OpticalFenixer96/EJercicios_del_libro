# a) Demostrar que la regla de actualizacin de pesos en la Ecuacin (1.3) se mueve en la direccin de clasificar correctamente x(t).

Se puede examinar la regla de actualizacin de pesos y ver cmo afecta la clasificacin de x(t).

La regla de actualizacin de pesos se expresa como:
w(t+1) = w(t) + y(t) * x(t)

Suponiendo que x(t) est clasificado incorrectamente por w(t), lo que significa que y(t) * w(t) * x(t)  0. Esto implicaria que la prediccin actual de y(t) sea incorrecta.

Ahora considerando el vector de pesos actualizado w(t+1) y su producto escalar con x(t):

w(t+1) * x(t) = (w(t) + y(t) * x(t)) * x(t)
= w(t) * x(t) + y(t) * x(t) * x(t)

Dado que x(t) * x(t) siempre es positivo, tenemos:
w(t+1) * x(t) = w(t) * x(t) + y(t) * ||x(t)||^2

Enfocandonos en y(t) * ||x(t)||^2, se evidencia que representa un factor de escala positivo de x(t) basado en su norma al cuadrado, este trmino afecta la direccin de la actualizacin de pesos.

Si x(t) est clasificado incorrectamente por w(t), significa que y(t) * w(t) * x(t)  0, por ende el trmino de actualizacin y(t) * ||x(t)||^2 ser positivo y asi el vector de pesos se mover en la direccin de x(t), intentando corregir la clasificacin incorrecta.

Se concluye que la regla de actualizacin de pesos en la Ecuacin (1.3)  se mueve en la direccin de clasificar correctamente x(t) cuando x(t) est clasificado incorrectamente por w(t), como se muestra arriba.
